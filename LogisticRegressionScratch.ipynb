```python

import os

def read_all_files_in_directory(directory_path):
    """
    Reads the content of all files in a specified directory.

    Args:
        directory_path (str): The path to the directory to read files from.

    Returns:
        dict: A dictionary where keys are filenames and values are their content.
              Returns an empty dictionary if the directory does not exist or
              contains no readable files.
    """
    file_contents = {}
    datasetimages=np.zeros((12499,400));
    c=0;
    ##datasetimages=np.zeroes(12,499,)
    if not os.path.isdir(directory_path):
        print(f"Error: Directory '{directory_path}' does not exist.")
        return file_contents

    for filename in os.listdir(directory_path):
        filepath = os.path.join(directory_path, filename)
        if os.path.isfile(filepath):
            try:
                
                im=mpimg.imread(filepath);
                print(im.shape);
                im_flattened=im.flatten();
                print(im_flattened.shape);
                ##padwidth=800000-im_flattened.shape[0];
                ##im_flattened=np.pad(im_flattened,pad_width=padwidth,mode="constant",constant_values=0);
                im_flattened=im_flattened[0:400];
                datasetimages[c,]=im_flattened;
                
                """with open(filepath, 'r', encoding='utf-8') as f:
                    content = f.read()
                    file_contents[filename] = content"""
            except Exception as e:
                print(f"Could not read file '{filename}': {e}")
        c=c+1;
        print("Read",c,"images upto now");
    return datasetimages;

# Example usage:
target_directory = "C:/Users/parth/Downloads/archive(2)/PetImages/Cat"; # Replace with your directory path
all_files_data = read_all_files_in_directory(target_directory);

target_directory2="C:/Users/parth/Downloads/archive(2)/PetImages/Dog";
all_files_data2=read_all_files_in_directory(target_directory2);

"""for filename, content in all_files_data.items():
    print(f"--- Content of {filename} ---")
    print(content)
    print("-" * 30)"""
```

```python
class Logistic_Regression:

   

    def __init__(self):
        self.w=np.zeros(400);
        self.b=1;
        pass;
        
        
    def sigmoid_compute(self,x):
        x=np.transpose(x);
        dot_prod=np.dot(x,self.w);
        z=dot_prod+self.b;
        ##print(z.shape);
        z=float(z);
        return 1/(1+math.exp(-z));
       

    def binary_cross_entropy_loss(self,x,y):
        z1=self.sigmoid_compute(x);
        ##print(z.shape);
        s=-1*(y*math.log(z1,2)+(1-y)*math.log(1-z1,2));
        return s;

    def summed_loss(self,X,Y,n):
        s=0;
        for i in range(n):
            s+=self.binary_cross_entropy_loss(X[i,],Y[i]);
        return s*(1/12499);

    def grads(self,X,Y):
        w_grad=0;
        b_grad=0;
        c=0;
        for i in range(12499):
            c=self.sigmoid_compute(X[i,])-Y[i]
            w_grad+=X[i,]*c;
            b_grad+=c;
        return (w_grad,b_grad);

    def model_train(self,X,Y,e,lr,reg):
        w_t=np.zeros(400);
        b_t=0;
        sub=0;
        fac=(lr/12499);
        for i in range(e):
            wg,bg=self.grads(X,Y);
            w_t=w_t*(1-2*reg*lr)-lr*(1/12499)*wg;
            b_t=b_t-lr*(1/12499)*bg;
            self.w=w_t;
            self.b=b_t;
            print("In epoch: ",i+1, "b=",self.b," loss=",self.summed_loss(X,Y,12499));
    
    def model_test(self,X1,Y1):
        print("Loss = ",self.summed_loss(X1,Y1,X1[1,].shape[0]));







    
        
```
